### Read everthing in as pd.DataFrame

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
FOLDER_PATH = './results/'

# --------- baseline method --------
baseline_df_1 = pd.read_pickle(FOLDER_PATH + 'baseline_run0.pkl').sort_values('Instance')
baseline_df_2 = pd.read_pickle(FOLDER_PATH + 'baseline_run1.pkl').sort_values('Instance')
baseline_df_3 = pd.read_pickle(FOLDER_PATH + 'baseline_run2.pkl').sort_values('Instance')

# -------- greedy method ---------
greedy_df_1 = pd.read_pickle(FOLDER_PATH + 'naive_run0.pkl').sort_values('Instance')
greedy_df_2 = pd.read_pickle(FOLDER_PATH + 'naive_run1.pkl').sort_values('Instance')
greedy_df_3 = pd.read_pickle(FOLDER_PATH + 'naive_run2.pkl').sort_values('Instance')

# -------- jump-half-greedy --------
hybrid_jump_half_greedy_df_1 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_half_naive_run0.pkl').sort_values('Instance')
hybrid_jump_half_greedy_df_2 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_half_naive_run1.pkl').sort_values('Instance')
hybrid_jump_half_greedy_df_3 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_half_naive_run2.pkl').sort_values('Instance')

# -------- jump-half-exhaustive --------
hybrid_jump_half_exhaustive_df_1 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_half_exhaustive_run0.pkl').sort_values('Instance')
hybrid_jump_half_exhaustive_df_2 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_half_exhaustive_run1.pkl').sort_values('Instance')
hybrid_jump_half_exhaustive_df_3 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_half_exhaustive_run2.pkl').sort_values('Instance')

# -------- jump-std-exhaustive --------
hybrid_jump_std_exhaustive_df_1 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_std_exhaustive_run0.pkl').sort_values('Instance')
hybrid_jump_std_exhaustive_df_2 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_std_exhaustive_run1.pkl').sort_values('Instance')
hybrid_jump_std_exhaustive_df_3 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_std_exhaustive_run2.pkl').sort_values('Instance')

# -------- jump-3std-exhaustive --------
hybrid_jump_3std_exhaustive_df_1 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_3std_exhaustive_run0.pkl').sort_values('Instance')
hybrid_jump_3std_exhaustive_df_2 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_3std_exhaustive_run1.pkl').sort_values('Instance')
hybrid_jump_3std_exhaustive_df_3 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_3std_exhaustive_run2.pkl').sort_values('Instance')

# -------- jump-5std-exhaustive --------
hybrid_jump_5std_exhaustive_df_1 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_5std_exhaustive_run0.pkl').sort_values('Instance')
hybrid_jump_5std_exhaustive_df_2 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_5std_exhaustive_run1.pkl').sort_values('Instance')
hybrid_jump_5std_exhaustive_df_3 = pd.read_pickle(FOLDER_PATH + 'hybrid_jump_5std_exhaustive_run2.pkl').sort_values('Instance')
```

```python
# the dictionary can then be used to iterate for comparison.
df_dict = {}

# -------- baseline method --------
df_dict['baseline_df_1'] = baseline_df_1
df_dict['baseline_df_2'] = baseline_df_2
df_dict['baseline_df_3'] = baseline_df_3

# -------- greedy method ---------
df_dict['greedy_df_1'] = greedy_df_1
df_dict['greedy_df_2'] = greedy_df_2
df_dict['greedy_df_3'] = greedy_df_3

# -------- jump-half-greedy --------
df_dict['hybrid_jump_half_greedy_df_1'] = hybrid_jump_half_greedy_df_1
df_dict['hybrid_jump_half_greedy_df_2'] = hybrid_jump_half_greedy_df_2
df_dict['hybrid_jump_half_greedy_df_3'] = hybrid_jump_half_greedy_df_3

# -------- jump-half-exhaustive --------
df_dict['hybrid_jump_half_exhaustive_df_1'] = hybrid_jump_half_exhaustive_df_1
df_dict['hybrid_jump_half_exhaustive_df_2'] = hybrid_jump_half_exhaustive_df_2
df_dict['hybrid_jump_half_exhaustive_df_3'] = hybrid_jump_half_exhaustive_df_3

# -------- jump-std-exhaustive --------
df_dict['hybrid_jump_std_exhaustive_df_1'] = hybrid_jump_std_exhaustive_df_1
df_dict['hybrid_jump_std_exhaustive_df_2'] = hybrid_jump_std_exhaustive_df_2
df_dict['hybrid_jump_std_exhaustive_df_3'] = hybrid_jump_std_exhaustive_df_3

# -------- jump-3std-exhaustive --------
df_dict['hybrid_jump_3std_exhaustive_df_1'] = hybrid_jump_3std_exhaustive_df_1
df_dict['hybrid_jump_3std_exhaustive_df_2'] = hybrid_jump_3std_exhaustive_df_2
df_dict['hybrid_jump_3std_exhaustive_df_3'] = hybrid_jump_3std_exhaustive_df_3

# -------- jump-5std-exhaustive --------
df_dict['hybrid_jump_5std_exhaustive_df_1'] = hybrid_jump_5std_exhaustive_df_1
df_dict['hybrid_jump_5std_exhaustive_df_2'] = hybrid_jump_5std_exhaustive_df_2
df_dict['hybrid_jump_5std_exhaustive_df_3'] = hybrid_jump_5std_exhaustive_df_3
```

#### The total running time
Some observation from the running time are:
- The hybrid model is still slower than the baseline model;
- hybrid model is slower than naive model for ~2000 seconds which is probably because the naive model does not have probing phase;
- with the optimal prediction hybrid model is still slower than the naive model for ~2786 seconds, which once again might be because of the lack of probing phase.

## Runtime distribution

```python
plt.figure(figsize=(12,8))
# sns.distplot(baseline_df['runtime_ms'], hist = False, kde = True, kde_kws = {'linewidth': 2})
for key, val in df_dict.items():
#     print("Total running time of the {:<30}  model is  {:>10} seconds".format(key, int(val['runtime_ms'].sum() / 1000)))
    sns.kdeplot(val['runtime_ms'], shade=True, label=key)

# Plot formatting
plt.legend(prop={'size': 16}, title = 'runtime_ms')
plt.title('Density Plot of running time in ms')
plt.xlabel('rumtime_ms')
plt.ylabel('Density')
```

```python
pd.options.mode.chained_assignment = None  # default='warn'

to_visualize_df = pd.DataFrame(columns=['Instance', 'runtime_ms', 'type'])

for key, val in df_dict.items():
    temp_df = val[['Instance', 'runtime_ms']]
    temp_df['type'] = str(key[:-5])

    to_visualize_df = pd.concat([to_visualize_df, temp_df])
    
to_visualize_df['runtime_ms_log'] = np.log(to_visualize_df['runtime_ms'])
```

```python
to_visualize_df
```

```python
plt.figure(figsize=(26,8))

sns.boxplot(x='type', y='runtime_ms_log', data=to_visualize_df)
```

```python
sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})

# Initialize the FacetGrid object
pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
g = sns.FacetGrid(to_visualize_df, row="type", hue="type", aspect=15, height=2, palette=pal)

# Draw the densities in a few steps
g.map(sns.kdeplot, "runtime_ms", clip_on=False, cumulative=True, shade=True, alpha=1, lw=1.5, bw_method=.2)
# g.map(sns.kdeplot, "runtime_ms", clip_on=False, color="w", lw=2, bw=.2)
g.map(plt.axhline, y=0, lw=2, clip_on=False)


# Define and use a simple function to label the plot in axes coordinates
def label(x, color, label):
    ax = plt.gca()
    ax.text(0, .2, label, fontweight="bold", color=color,
            ha="right", va="center", transform=ax.transAxes)


g.map(label, "runtime_ms")

# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-.25)

# Remove axes details that don't play well with overlap
g.set_titles("")
g.set(yticks=[])
g.despine(bottom=True, left=True)
```

```python
ordered_types = list(to_visualize_df['type'].value_counts().index).sort()
g = sns.FacetGrid(to_visualize_df, row="type", row_order=ordered_types,
                  height=1.7, aspect=4,)
# g.map(sns.distplot, "runtime_ms", hist=False, rug=False);
# g.map(sns.kdeplot, "runtime_ms_log", cumulative=True);
g.map(sns.kdeplot, "runtime_ms", cumulative=True);
```
